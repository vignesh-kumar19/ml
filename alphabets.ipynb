```
# https://machinelearningmastery.com/understanding-stateful-lstm-recurrent-neural-networks-python-keras/
from keras.layers import LSTM, Dense
from keras.models import Sequential
import numpy as np
from keras.utils.np_utils import to_categorical
from keras.preprocessing.sequence import pad_sequences
```

```
# Creating X and Y
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

char_to_int = dict([(ch, i) for i, ch in enumerate(alphabet)])
int_to_char = dict([(i, ch) for i, ch in enumerate(alphabet)])

rawX = list()
rawY = list()
max_sequence = 5
samples = 1000
for _ in range(samples):
    start = np.random.randint(len(alphabet)-2)
    end = np.random.randint(start, min(start+max_sequence, len(alphabet) -1))
    
    rawX.append([char_to_int[ch] for ch in alphabet[start:end+1]])
    rawY.append(char_to_int[alphabet[end+1]])

X = pad_sequences(sequences=rawX, maxlen=max_sequence, dtype='float32') / len(alphabet)
Y = to_categorical(rawY)
```

```
print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```
#  Reshape X as (Num of samples, sequences, features)
X = np.reshape(X, newshape=(X.shape[0], X.shape[1], 1))
```

```
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(Y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, Y, batch_size=1, epochs=500)
```

```
rawXtest = []
for _ in range(10):
  start = np.random.randint(len(alphabet)-2)
  end = np.random.randint(start, min(start+max_sequence, len(alphabet)-1))

  rawXtest.append([char_to_int[ch] for ch in (alphabet[start:end])])
Xtest = pad_sequences(rawXtest, maxlen=max_sequence)
Xtest = Xtest / len(alphabet)
Xtest = np.reshape(Xtest, newshape=(Xtest.shape[0],Xtest.shape[1], 1))
predtest = model.predict(Xtest, verbose=0)
predtest = np.argmax(predtest, axis=1)

for i in range(len(rawXtest)):
  seq_in = [int_to_char[0] if no == None else int_to_char[no]  for no in rawXtest[i]]
  seq_out = int_to_char[predtest[i]]
  
  print(seq_in, '-->', seq_out)
```
